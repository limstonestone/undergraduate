```python
import pandas as pd
import matplotlib.pyplot as plt
import FinanceDataReader as fdr
fdr.__version__
```

```python
df = fdr.DataReader("005930", "2020")
df.tail(10) 

# 삼성전자: 005930
# 현대차: 005380
# 엔씨소프트: 036570
# 셀트리온: 068270
# 한국항공우주: 047810
# LG에너지솔루션: 373220
# 카카오: 035720
# S-Oil: 010950
# 고려아연: 010130
# 롯데쇼핑: 023530
# CJ제일제당: 097950
# HMM: 011200
# 두산에너빌리티: 034020
# 삼성생명: 032830
```

```python
df_cp = df["Close"]
df["Return_iloc"] = ( df_cp/df_cp.iloc[0] - 1 ) # 수익률
df
```

```python
g = df["Return_iloc"].plot(figsize=(10, 4))
```

```python
df["Return_daily"] = ( df_cp/df_cp.shift(1) - 1 )   # 일별 수익률 (1차 차분)
df
```

```python
g = df["Return_daily"].plot(figsize=(10, 4))
```

```python
df["Return_pct_change"] = df_cp.pct_change()    # 일별 수익률과 동일한 Method
df
```

```python
g = df["Return_pct_change"].plot(figsize=(10, 4))
```

```python
# 누적수익률
df["Return_daily_cum"] = ( df["Return_daily"] + 1 ).cumprod()
g = df["Return_daily_cum"].plot(figsize=(10, 4))
g = plt.axhline(1, c="k", lw=1) 
```

```python
# 이동평균
df["MA_5"] = df_cp.rolling(5).mean()    # 5 영업일
df["MA_10"] = df_cp.rolling(10).mean()  # 10 영업일
df["MA_20"] = df_cp.rolling(20).mean()  # 20 영업일
```

```python
g = df[["Close", "MA_5", "MA_10", "MA_20"]].plot(figsize=(10, 4), title="Moving Average")
```

```python
# MA : 특정한 기간 동안의 수익률
# CMA(Cumulative MA) : 첫 번째 고려되는 윈도우가 1~5 영업일 / 누적평균 수익률

df["CMA_5"] = df_cp.expanding(5).mean()
```

```python
g = df[["Close", "CMA_5"]].plot(figsize=(10,4), title="Cumulative moving average")
```

```python
# 지수이동평균

df["EWM_short"] = df_cp.ewm(span=12).mean() # 단기에서는 보통 12
df["EWM_long"] = df_cp.ewm(span=26).mean()  # 장기에서는 보통 26
```

```python
g = df[["Close", "EWM_short", "EWM_long"]].plot(figsize=(10, 4), title="Exponential Moving Average")
```

```python
g = df[["Close", "MA_5", "CMA_5", "EWM_long"]].plot(figsize=(10, 4))
```
